```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
    
import warnings
warnings.filterwarnings('ignore')
```

```python
ipl = pd.read_csv("ipl_2022_dataset.csv")
ipl.head()
```

```python
ipl.shape
```

```python
ipl.info()
```

```python
ipl.columns
```

```python
ipl.drop("Unnamed: 0",axis =1,inplace =True)
ipl.head()
```

```python
ipl.isnull().sum()
```

```python
ipl[ipl['Cost IN $ (000)'].isnull()]
```

```python
ipl['Cost IN $ (000)'] = ipl['Cost IN $ (000)'].fillna(0)
ipl['COST IN ₹ (CR.)'] = ipl['COST IN ₹ (CR.)'].fillna(0)
```

```python
ipl[ipl['2021 Squad'].isnull()]
```

```python
ipl['2021 Squad'] = ipl['2021 Squad'].fillna('Not Participated')
```

```python
ipl.isnull().sum()
```

```python
teams = ipl[ipl["COST IN ₹ (CR.)"]>0]['Team'].unique()
teams
```

```python
ipl['status'] = ipl['Team'].replace(teams,'sold')
```

```python
ipl.head()
```

```python
ipl[ipl['Player'].duplicated(keep =False)]
```

```python
#how many player have participated in 2022 IPL auction?
ipl.shape[0]
```

```python
types = ipl['TYPE'].value_counts()
types.reset_index()
```

```python
plt.pie(types.values,labels=types.index,labeldistance=1.2,autopct='%1.2f%%', shadow =True,startangle=60)
plt.title("Role of Players participated",fontsize = 15)
plt.show()
```

```python
squad_2021 = ipl['2021 Squad'].value_counts()
squad_2021.reset_index()
# types
```

```python
plt.pie(squad_2021.values,labels=squad_2021.index,labeldistance=1.2,autopct='%1.2f%%', shadow =True,startangle=60)
plt.title("2021 paticipation and Result",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (10,5))
fig = sns.countplot(x = 'status',data = ipl,palette = ['Green', 'Red'])
plt.xlabel('Sold or Unsold')
plt.ylabel('Number of Players')
plt.title('Sold vs Unsold',fontsize =15)
plt.plot()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.0f'), (p.get_x() +
                 p.get_width()/2., p.get_height()), ha = 'center',
                 va = 'center', xytext = (0,4), textcoords = 'offset points')
print(fig.patches)
```

```python
ipl.groupby('status')['Player'].count()
```

```python
data = ipl[ipl['Team']!='Unsold']['Team']
data.reset_index()
c = data.values
data = pd.DataFrame(columns = ['value'])
data['value'] = c
data.head(8)
```

```python
plt.figure(figsize=(20,10))
fig= sns.countplot(data = data ,x='value')
plt.xlabel("Team Names")
plt.ylabel("Number of Players")
plt.title("Players Bought by Each Team",fontsize= 15)
plt.xticks(rotation =70)
plt.plot()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.0f'), (p.get_x() +
                 p.get_width()/2., p.get_height()), ha = 'center',
                 va = 'center', xytext = (0,4), textcoords = 'offset points')
```

```python
ipl['retention'] = ipl['Base Price']
```

```python
ipl['retention'].replace(['2 Cr' ,'1 Cr','1.5 Cr', '40 Lakh', '20 Lakh','75 Lakh','50 Lakh','30 Lakh']
                          , 'From Auction', inplace =True)
```

```python
ipl['Base Price'].replace('Draft Pick',0,inplace = True)
```

```python
ipl["base_price_unit"] = ipl["Base Price"].apply(lambda x: str(x).split(' ')[-1])
ipl['base_price'] = ipl['Base Price'].apply(lambda x: str(x).split(' ')[0])
```

```python
ipl['base_price'].replace('Retained',0,inplace =True)
```
